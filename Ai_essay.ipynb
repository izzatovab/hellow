```python
# 1. Ma'lumotlarni yuklab olish (UCI Wine Quality dataset)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
data = pd.read_csv(url, delimiter=';')
```

```python
# 2. Ma'lumotlarni tekshirish va tayyorlash
print(data.head())  # Ma'lumotlar strukturasini ko'rish
print(data.isnull().sum())  # Yetishmayotgan qiymatlar yo'q

# Regression uchun tayyorlash (Quality - nuqtali baho)
X_reg = data.drop('quality', axis=1)
y_reg = data['quality']

# Train-test split (Regression)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)
```

```python
# 3. Oddiy Chiziqli Regressiya (Bitta xususiyat bilan)
# "alcohol" xususiyatini tanlaymiz
X_train_simple = X_train_reg[['alcohol']]  # Faqat train ma'lumotlari
model_simple = LinearRegression()
model_simple.fit(X_train_simple, y_train_reg)

# Bashorat va baholash (TEST ma'lumotlari bilan)
X_test_simple = X_test_reg[['alcohol']]    # Faqat test ma'lumotlari
y_pred_simple = model_simple.predict(X_test_simple)  # <-- Xatolik shu yerda edi

print("\nOddiy Chiziqli Regressiya Metrikalari:")
print(f"MAE: {mean_absolute_error(y_test_reg, y_pred_simple):.2f}")
print(f"MSE: {mean_squared_error(y_test_reg, y_pred_simple):.2f}")
print(f"R²: {r2_score(y_test_reg, y_pred_simple):.2f}")
```

```python
# 4. Ko'p O'zgaruvchili Chiziqli Regressiya
model_multi = LinearRegression()
model_multi.fit(X_train_reg, y_train_reg)

y_pred_multi = model_multi.predict(X_test_reg)
print("\nKo'p O'zgaruvchili Regressiya Metrikalari:")
print(f"MAE: {mean_absolute_error(y_test_reg, y_pred_multi):.2f}")
print(f"MSE: {mean_squared_error(y_test_reg, y_pred_multi):.2f}")
print(f"R²: {r2_score(y_test_reg, y_pred_multi):.2f}")
```

```python
# 5. Logistik Regressiya uchun ma'lumotlarni tayyorlash
# Qualityni ikkita klassga ajratamiz (>=6 yoki <6)
data['quality_class'] = np.where(data['quality'] >= 6, 1, 0)
X_clf = data.drop(['quality', 'quality_class'], axis=1)
y_clf = data['quality_class']

# Train-test split (Classification)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)

# Modelni yaratish va o'qitish
model_clf = LogisticRegression(max_iter=1000)
model_clf.fit(X_train_clf, y_train_clf)

# Baholash
y_pred_clf = model_clf.predict(X_test_clf)
print("\nLogistik Regressiya Metrikalari:")
print(f"Aniqlik: {accuracy_score(y_test_clf, y_pred_clf):.2f}")
print("Classification Report:\n", classification_report(y_test_clf, y_pred_clf))
print("Confusion Matrix:\n", confusion_matrix(y_test_clf, y_pred_clf))
```
